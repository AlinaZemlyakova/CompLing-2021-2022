```python
import re

def remove_tags(text):
  #remove tags
  res = re.sub('\<[^>]*\>', '', text)
  #remove merge
  res = re.sub('([.?!])([А-Я0-9]+)', '\g<1> \g<2>', res)
  res = re.sub('\n|\r', '', res)
  return res

text = remove_tags(text)
print (text)
```
